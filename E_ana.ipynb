```python
import numpy as np
import matplotlib.pyplot as plt

f = np.loadtxt('/Users/keitatanaka/Dropbox/share/work/microcalorimeters/experiment/COMSOL/Energy_Conservation/run0.txt')
t = f[:,0]
Tsou = f[:,1]
Ttes = f[:,2]
Tmem = f[:,3]
I = f[:,4]
Tabs = f[:,5]
dTtes = np.diff(Ttes)
dTabs = np.diff(Tabs)
dTmem = np.diff(Tmem)
Cabs = 0.825e-12
Ctes = 0.048e-12
Cmem = 2.3571e-14
JtoeV = 1/1.60217662e-19
```

```python
#plt.plot(t, Tsou, label='Tsou')
plt.plot(t, Ttes, label='Ttes')
#plt.plot(t, Tmem, label='Tmem')
plt.plot(t, Tabs, label='Tabs')
plt.legend()
plt.show()
```

```python
plt.plot(t[1:],dTabs*Cabs*JtoeV, label='dTabs')
plt.plot(t[1:],dTtes*Ctes*JtoeV, label='dTtes')
plt.plot(t[1:],dTmem*Cmem*JtoeV, label='dTmem')
q = dTtes*Ctes + dTabs*Cabs + dTmem*Cmem
plt.plot(t[1:],q*JtoeV*1e+3, label='q')
plt.xlim(1.0249,1.0254)
plt.legend()
plt.xscale('log')

plt.yscale('log')
plt.show()
```
