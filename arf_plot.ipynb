```python
from xspec import *
import os
from xrism_tools import XspecFit
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt

f = fits.open('/Users/keitatanaka/Downloads/fake_spectrum/fakeit_result_ratio/fakeit_spec_8_b1.pi')
counts = f[1].data['COUNTS']
channel = f[1].data['CHANNEL'] / 2000

plt.step(channel, counts, where='mid', label='fake spectrum', color='black')
plt.xlim(5.9,6.2)
cnt_mask = (channel > 6.0) & (channel < 6.1)
cnt_sum = np.sum(counts[cnt_mask])
print(f"Counts in the range 6.0-6.1 keV: {cnt_sum}")
```
