```python
import numpy as np
import matplotlib.pyplot as plt

class ContourPlotter:
    def __init__(self, delta, x_params, y_params):
        self.delta_1d = np.array(delta)
        self.xnum, self.xmin, self.xmax, self.xstep = x_params
        self.ynum, self.ymin, self.ymax, self.ystep = y_params
        self.delta_2d = self.delta_1d.reshape((self.ystep + 1, self.xstep + 1))
        self.x_values = np.linspace(self.xmin, self.xmax, self.xstep + 1)
        self.y_values = np.linspace(self.ymin, self.ymax, self.ystep + 1)

    def plot_contours(self, X, Y, contour_levels):
        plt.contourf(X, Y, self.delta_2d, cmap='Greys')
        plt.colorbar(label='delta')
        return plt.contour(X, Y, self.delta_2d, levels=contour_levels, colors=['red', 'green', 'blue'])

    def find_intersections(self, contour_paths):
        intersections = []
        for path in contour_paths:
            vertices = path.vertices
            intersections.extend(vertices)
        return intersections

    def calculate_and_plot_slope(self, intersection_points, x_values, contour_level):
        if len(intersection_points) >= 2:
            first_point_by_increasing_slope = min(intersection_points, key=lambda p: p[1] / p[0] if p[0] != 0 else np.inf)
            first_point_by_decreasing_slope = max(intersection_points, key=lambda p: p[1] / p[0] if p[0] != 0 else np.inf)

            x1, y1 = first_point_by_increasing_slope
            x2, y2 = first_point_by_decreasing_slope

            # Calculate slopes
            if x1 != 0:
                a1 = y1 / x1
            else:
                a1 = np.inf  # Avoid division by zero
            if x2 != 0:
                a2 = y2 / x2
            else:
                a2 = np.inf  # Avoid division by zero
            
            print(f"Contour Level {contour_level}: Calculated slopes a: {a1:.2f}, {a2:.2f}")

            # Plotting lines based on calculated slopes
            y_line1 = a1 * x_values
            y_line2 = a2 * x_values
            plt.plot(x_values, y_line1, label=f'y = {a1:.2f}x (level {contour_level})', color='red')
            plt.plot(x_values, y_line2, label=f'y = {a2:.2f}x (level {contour_level})', color='red')
        
            # Plot intersection points
            plt.plot(x1, y1, 'ro')
            plt.plot(x2, y2, 'ro')
            print(f"Contour Level {contour_level}: Intersection Points: ({x1}, {y1}), ({x2}, {y2})")

    def steppar(self, contour_levels):
        X, Y = np.meshgrid(self.x_values, self.y_values)
        contours = self.plot_contours(X, Y, contour_levels)

        for contour_level in contour_levels:
            intersection_points = []
            for contour in contours.collections:
                # Check if contour level matches the current contour level
                if contour_level in contour.get_array():
                    contour_paths = contour.get_paths()
                    intersection_points.extend(self.find_intersections(contour_paths))

            print(f"\nContour Level {contour_level}:")
            self.calculate_and_plot_slope(intersection_points, self.x_values, contour_level)

        plt.xlabel('X')
        plt.ylabel('Y')
        plt.xlim(self.xmin, self.xmax)
        plt.ylim(self.ymin, self.ymax)
        plt.legend()
        plt.title('Color Map with delta contours and y=ax lines')
        plt.show()

# Example usage
ynum = 26
ymin = 9E-5
ymax = 11E-5
ystep = 10
xnum = 22
xmin = 1E-5
xmax = 5.5E-5
xstep = 10
sigma = 1.0

# delta (example random data)
delta = np.array(delta) # Random delta values

x_params = (xnum, xmin, xmax, xstep)
y_params = (ynum, ymin, ymax, ystep)

plotter = ContourPlotter(delta, x_params, y_params)
contour_levels = [3, 5, 10]
plotter.steppar(contour_levels)

```

```python
import numpy as np
from astropy.constants import G, m_p, k_B
import matplotlib.pyplot as plt

# NFWパラメータ
rho_0 = 1e-3  # 特性密度 (kg/m^3) 例
r200 = 1e6  # 銀河団の特性半径 (kpc)
c = 5  # 集中度パラメータ
r_s = r200 / c  # スケール半径 (kpc)
f_b = 0.16  # バリオン質量分率
mu = 0.6  # 平均粒子質量

# 銀河団の静水圧平衡から温度分布を求めるためのNFWプロファイル
def nfw_density(r, rho_0, r_s):
    return rho_0 / ((r / r_s) * (1 + r / r_s)**2)

# 質量密度からバリオン密度を計算
def baryon_density(r, rho_0, r_s, f_b):
    return f_b * nfw_density(r, rho_0, r_s)

# バリオン密度から電子密度を計算
def electron_density(r, rho_0, r_s, f_b, mu):
    rho_b = baryon_density(r, rho_0, r_s, f_b)
    return rho_b / (mu * m_p.value)

# 温度分布を計算する関数 (単純な静水圧平衡モデル)
def temperature_profile(r, rho_0, r_s, f_b, mu):
    # 質量密度とポテンシャルを考慮
    # ここでは簡易的に比例関係を仮定
    # 実際には重力ポテンシャルの計算が必要
    return 1e7 * (r / r_s)**(-1)  # 温度のスケールを1e7Kとして仮定

# 半径の範囲 (kpc単位)
radii = np.linspace(1, 1000, 1000)

# 電子密度と温度の計算
electron_densities = electron_density(radii, rho_0, r_s, f_b, mu)
temperatures = temperature_profile(radii, rho_0, r_s, f_b, mu)

# プロット
fig, ax1 = plt.subplots()

color = 'tab:blue'
ax1.set_xlabel('Radius (kpc)')
ax1.set_ylabel('Electron Density (m^-3)', color=color)
ax1.plot(radii, electron_densities, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()
color = 'tab:red'
ax2.set_ylabel('Temperature (K)', color=color)
ax2.plot(radii, temperatures, color=color)
ax2.tick_params(axis='y', labelcolor=color)

plt.show()
```
